```python
df['Fecha_lanzamiento'] = pd.to_datetime(df['Fecha_lanzamiento'], errors='coerce')
df['Ao_lanzamiento'] = df['Fecha_lanzamiento'].dt.year
df['Mes_lanzamiento'] = df['Fecha_lanzamiento'].dt.month

def get_season(month):
    if 3 <= month <= 5:
        return 'Primavera'
    elif 6 <= month <= 8:
        return 'Verano'
    elif 9 <= month <= 11:
        return 'Otoo'
    else:
        return 'Invierno'

df['Estacion_lanzamiento'] = df['Mes_lanzamiento'].apply(get_season)

df.head()

df['Duracion_min'] = df['Duracion_mliseg'] / 1000 / 60
```

```python
X = df_clean.drop(columns=[ 'Popularidad', 'Mayor_popularidad' , 'Fecha_lanzamiento'])
y = df_clean['Mayor_popularidad']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# Alinear columnas en caso de que falte alguna categora en test o train
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Entrenamiento y prediccin con KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print("Exactitud:", round(accuracy_score(y_test, y_pred), 4))
print(classification_report(y_test, y_pred))
```

Matriz de confusion

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Baja", "Alta"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Matriz de Confusin")
plt.show()
```

Curva ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_prob = knn.predict_proba(X_test)[:, 1]  # Probabilidad de clase positiva
fpr, tpr, _ = roc_curve(y_test, y_prob)
auc = roc_auc_score(y_test, y_prob)

plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel("Tasa de falsos positivos")
plt.ylabel("Tasa de verdaderos positivos")
plt.title("Curva ROC")
plt.legend()
plt.grid(True)
plt.show()
```

Modelo 3: rboles de Decisin (CART), Poda, ROC y AUC

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
```

```python
X = df_clean[['Energia', 'Bailabilidad', 'Vivacidad', 'Valencia', 'Duracion_min', 'Acustico']]
y = df_clean['Mayor_popularidad']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
tree_full = DecisionTreeClassifier(random_state=42)
tree_full.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,10))
plot_tree(tree_full, feature_names=X.columns, class_names=['Baja_popularidad', 'Alta_popularidad'], filled=True, max_depth=3)
plt.title("rbol sin poda (visualizacin parcial)")
plt.show()
```

```python
y_pred_full = tree_full.predict(X_test)
print("Reporte de clasificacin - rbol sin poda")
print(classification_report(y_test, y_pred_full))
```

```python
tree_pruned = DecisionTreeClassifier(random_state=42, max_depth=4)
tree_pruned.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,8))
plot_tree(tree_pruned, feature_names=X.columns, class_names=['Baja_popularidad', 'Alta_popularidad'], filled=True)
plt.title("rbol con poda (completo)")
plt.show()
```

```python
y_pred_pruned = tree_pruned.predict(X_test)
print("Reporte de clasificacin - rbol podado")
print(classification_report(y_test, y_pred_pruned))
```

```python
#Matriz de confusion

ConfusionMatrixDisplay.from_estimator(tree_pruned, X_test, y_test)
plt.title("Matriz de Confusin - rbol Podado")
plt.show()
```

```python
#Curva Roc

y_scores = tree_pruned.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"ROC (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Aleatorio")
plt.xlabel("Tasa de Falsos Positivos")
plt.ylabel("Tasa de Verdaderos Positivos")
plt.title("Curva ROC - rbol Podado")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
